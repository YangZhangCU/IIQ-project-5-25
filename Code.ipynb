```python
#read the excel
import openpyxl
wb = openpyxl.load_workbook('Raw Data.xlsx')
sheet = wb.get_sheet_by_name('Sheet1')
```

```python
#create the dict
data_2014 = dict()
data_2013 = dict()
data_2012 = dict()
data_2011 = dict()
data_2010 = dict()
data_2009 = dict()
data_2008 = dict()
data_2007 = dict()
data_2006 = dict()
data_2005 = dict()
```

```python
#2014
for i in range(1,1200,4):
    data_2014[sheet.cell(row=i+2, column=0).value]=[sheet.cell(row=i+1, column=0).value, sheet.cell(row=i, column=0).value, sheet.cell(row=i+3, column=0).value]
#2013
for i in range(1,1204,4):
    data_2013[sheet.cell(row=i+2, column=1).value]=[sheet.cell(row=i+1, column=1).value, sheet.cell(row=i, column=1).value, sheet.cell(row=i+3, column=1).value]
#2012
for i in range(1,1200,4):
    data_2012[sheet.cell(row=i+2, column=2).value]=[sheet.cell(row=i+1, column=2).value, sheet.cell(row=i, column=2).value, sheet.cell(row=i+3, column=0).value]
```

```python
#2011
for i in range(1,302):
    data_2011[sheet.cell(row=i, column=4).value]=[sheet.cell(row=i, column=3).value, sheet.cell(row=i, column=5).value, sheet.cell(row=i, column=6).value]
```

```python
#generate the organization list
organization = set(sorted(data_2005.keys())) | set(sorted(data_2006.keys())) | set(sorted(data_2007.keys())) | set(sorted(data_2008.keys())) | set(sorted(data_2009.keys())) | set(sorted(data_2010.keys()))  | set(sorted(data_2011.keys())) | set(sorted(data_2012.keys())) | set(sorted(data_2013.keys())) | set(sorted(data_2014.keys()))
organization = sorted(list(organization))
len(organization)
```

```python
#write it to csv
outcome = list()
data = [data_2014, data_2013,data_2012,data_2011,data_2010,data_2009,data_2008,data_2007,data_2006,data_2005,]
outcome.append(list(['Rank_2014','Patents_2014','Change_2014','Rank_2013','Patents_2013','Change_2013','Rank_2012','Patents_2012','Change_2012','Rank_2011','Patents_2011','Change_2011','Rank_2010','Patents_2010','Rank_2009','Patents_2009','Rank_2008','Patents_2008','Rank_2007','Patents_2007','Rank_2006','Patents_2006','Rank_2005','Patents_2005']))
for i in range(770):
    temp = list()
    for j in range(4):
        if (organization[i] in data[j]):
            temp.extend(data[j][organization[i]])
        else:
            temp.extend([0,0,0])    
    for j in range(6):
        if (organization[i] in data[j+4]):
            temp.extend(data[j+4][organization[i]])
        else:
            temp.extend([0,0])
    outcome.append(temp)
import csv
with open('Top300 Patent Companies.csv', 'w', newline='') as csvfile:
    spamwriter = csv.writer(csvfile, delimiter=',',quotechar='|', quoting=csv.QUOTE_MINIMAL)
    for j in range(len(outcome)):
        spamwriter.writerow(outcome[j])
```
